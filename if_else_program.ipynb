```python
#Salary and Bonus
salary = float(input("Enter Salary:"))
experience = int(input("Enter Experience:"))
if experience == 5 :
    total = salary + (salary*(5/100))
    print("Bonus :",total)
elif experience == 10:
    total = salary +(salary*(10/100))
    print("Bonus:",total)
else:
    print("No Bonus")
```

```python
#Quandrant Program
x = int(input("Enter x axis value:"))
y = int(input("Enter Y axis value"))
if x>0 and y>0:
    print(f"{x} {y} is in Quadrant1")
elif x<0 and y>0:
        print(f"{x} {y} is in Quadrant2")
elif x<0 and y<0:
        print(f"{x} {y} is in Quadrant3")
elif x>0 and y>0:
        print(f"{x} {y} is in Quadrant4")
elif x==0 and y==0:
        print(f"{x} {y} is in Center")
```

```python
#Unit Calculation Program
unit = int(input("Enter Units:"))
if unit < 100:
    print("total Rate",unit*2)
elif unit >=100 and unit<=300:
    print("total Rate",unit*4)
elif unit >300 :
    print("total Rate",unit*9)
else:
    print("Invalid Input..")
    
```
